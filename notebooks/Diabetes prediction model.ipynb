## Project - Diabetes Prediction Using Machine Learning

## Problem Statement

Early detection of diabetes plays a critical role in preventing long-term complications. However, many individuals remain undiagnosed due to limited screening. This project aims to develop a machine learning model that predicts whether an individual is diabetic based on key medical and lifestyle indicators.

## Objective

- Analyze medical features to understand factors influencing diabetes.

- Build a reliable classification model to predict diabetes status.

- Evaluate model performance using accuracy and other metrics.

- Enable early risk identification using a simple, data-driven ML approach.

```python
#importing the dependencies 

import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import train_test_split 
from sklearn import svm
from sklearn.metrics import accuracy_score
```

### Data Collection & Analysis 

```python
#loading the diabetes dataset to a pandas data frame 

df = pd.read_csv("diabetes.csv",lineterminator="\n")
df
```

**Dataset Information**

##### The dataset contains 768 records with the following features:

- Pregnancies – number of times the individual has been pregnant
- Glucose – plasma glucose concentration during an oral glucose tolerance test
- BloodPressure – diastolic blood pressure (mm Hg)
- SkinThickness – triceps skinfold thickness (mm)
- Insulin – 2-hour serum insulin (mu U/ml)
- BMI – Body Mass Index (weight in kg / height in m²)
- DiabetesPedigreeFunction – likelihood of diabetes based on family history
- Age – age of the individual
- Outcome – target variable (1 = diabetic, 0 = non-diabetic)

### Splitting the Features and Target

```python
#separating the data and labels

X = df.drop(columns = "Outcome", axis = 1)
Y = df["Outcome"]
```

```python
print(X)
```

```python
print(Y)
```

### Data Standardization 

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
print(X_scaled)
```

```python
#saving the variable named X 

X = X_scaled
```

### Splitting the data into Training and Testing Part

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

### Training the Model

```python
model = svm.SVC(kernel="linear")
```

```python
#training the support vector machine classifier 

model.fit(X_train,Y_train)
```

### Model Evaluation 

```python
#accuracy score on the training data 

X_train_pred = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_pred,Y_train)
```

```python
print("Training data accuracy score: ",training_data_accuracy)
```

```python
#accuracy score on the testing data 

X_test_pred = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_pred,Y_test)
```

```python
print("Testing data accuracy score: ",testing_data_accuracy)
```

### Making a Predictive System 

```python
input_data = (4,129,86,20,270,35.1,0.231,23)

#changing the input data to numpy array
input_data_numpy_array = np.asarray(input_data)

#reshape the numpy array as we are predicting for one instance
input_data_reshaped = input_data_numpy_array.reshape(1,-1)

#standardize the input data 
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = model.predict(std_data)
print(prediction)


if (prediction[0] == 1):
    print("The person is Diabetic.")
else:
    print("The person is not Diabetic.")
```

### Conclusion 

**Key Insights**

- Glucose and BMI show the strongest association with diabetes.
- Older individuals tend to have a higher chance of being diabetic.
- Zero values in medical features must be treated before training the model.
